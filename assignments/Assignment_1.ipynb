**Answer Q2.11 Written Here:**

12. Create two Python functions that take a Numpy array (`x`) and scalar alpha ($\alpha$) as inputs, perform the mathematical formulas outlined below, and return the resulting Numpy array as an output. These two functions should have the names `F` and `G` so that you can call them as `F(x, alhpha)` and `G(x, alpha)`, respectively. Evaluate both functions at each point on your created grid. Call their outputs `out_F` and `out_G`, respectively. Start with a value of $\alpha = 1$ for both evaluations.

13. Plot the values of the outputs from both functions in the same plot. Include the same x and y axis labels as before, include a title that describes what the plot is showing, and label the data as `F(x)` and `G(x)` in the legend.

\begin{gather}
  \\
  \textbf{Defintions of } F(x, \alpha) \textbf{ and } G(x, \alpha) \\
  F(x, \alpha) = \frac{\text{sin}(x + 2)}{\alpha} \\
  G(x, \alpha) = \alpha * \text{cos}(x + 3) \\
  \text{Note: } \alpha \text{ is written in text as "alpha"}
\end{gather}

```python
x_grid = np.linspace(-6.5, 6.5, 100)

def F(x, alpha):
    return np.sin(x + 2) / alpha

def G(x, alpha):
    return alpha * np.cos(x + 3)

out_F = F(x_grid, 1)
out_G = G(x_grid, 1)

plt.scatter(x_grid, out_F, label='F(x)')
plt.scatter(x_grid, out_G, label='G(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('F(x) and G(x) functions (Alpha=1)')
plt.legend()
plt.show()
```
